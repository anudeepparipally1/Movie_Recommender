```python
'''keywords from genres
The ast module helps Python applications to process trees of the Python abstract syntax grammar'''
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
#keywords from 
def convertcast(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c=c+1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convertcast)
```

```python
movies.head(2)
```

```python
def convert_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(convert_director)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
#removing the space between the names
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tages']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies.head(14)
```

```python
'''required data of id,title,tages'''
new_data=movies[['id',"original_title",'tages','rating']]
```

```python
new_data.head(2)
```

```python
#for testing
new_data[new_data['original_title']=='Avatar']['rating']
```

```python
new_data['tages']=new_data['tages'].apply(lambda x:" ".join(x))
```

```python
new_data.head(2)
```

```python
new_data['tages']=new_data['tages'].apply(lambda x:x.lower())
```

```python
new_data.head(2)
```

```python
'''T(NLTK) is a platform used for building Python programs that work with
human language data for applying in statistical natural language processing (NLP)'''

import nltk
```

```python
'''the Porter stemming algorithm (or 'Porter stemmer') is a process for removing the 
commoner morphological and inflexional endings from words in English'''

    from nltk.stem import PorterStemmer
    stemmer=PorterStemmer()
```

```python
def stem(text):
    k=[]
    for i in text.split():
        k.append(stemmer.stem(i))
    return " ".join(k)
```

```python
new_data['tages']=new_data['tages'].apply(stem)
```

```python
new_data.head(2)
```

```python
'''Scikit-learn's CountVectorizer is used to convert a collection of text documents to a vector of term/token counts.
It also enables the â€‹pre-processing of text data prior to generating the vector representation. 
This functionality makes it a highly flexible feature representation module for text.'''

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv= CountVectorizer(max_features=5000,stop_words='english')
tv= TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_data['tages']).toarray()
vectors2=tv.fit_transform(new_data['tages']).toarray()
```

```python
tv.fit_transform(new_data['tages']).toarray().shape
```

```python
cv.fit_transform(new_data['tages']).toarray().shape
```

```python
#cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Compute cosine similarity between samples in X and Y
similarity=cosine_similarity(vectors)
similarity2=cosine_similarity(vectors2)
```

```python
#recommend function is used printing the top 5 related movies
def recommend(movie):
    movie_index=new_data[new_data['original_title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    r=[]
    m=[]
    rt=[]
    k=1
    for i in movie_list:
        r.append(float(new_data[new_data['original_title']=="{}".format(new_data.iloc[i[0]].original_title)]['rating']))
        m.append(new_data.iloc[i[0]].original_title)
        print("{} ->{}".format(k,new_data.iloc[i[0]].original_title))
        k=k+1
    for i in range(10):
        
        rt.append(r[i])
        
    s = np.array(rt)
    c=1
    j=1
    sort_index = np.argsort(s)
    print("<------------------****Top 5 rated movies among the above 10 movies****------------------>")
    for i in sort_index[::-1]:
        if(c<=5):
            print("{} ->{}".format(j,m[i]))
            c=c+1
            j=j+1
    



```
